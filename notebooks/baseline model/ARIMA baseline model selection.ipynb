```python
%load_ext autoreload
%autoreload 2

from scripts.arima_import import *
from scripts.ARIMAForcaster import ARIMAForecaster

dl = DataLoader('../data/database.db')

# constant
order_range = range(1, 25)
```

## Wheat

```python
# wheat_us_srw
df = dl.get_commodity_historical_monthly_data('wheat_us_srw').dropna().reset_index(drop=True)
df.columns = ['date', 'wheat_us_srw']

model = ARIMAForecaster(df=df.copy(), column='wheat_us_srw', date_column='date', 
                        order=(1,0,0))
order_performance_df = model.evaluate_ar_values(start_date='2018-01-01', 
                                                end_date='2019-12-01',
                                                order_range=order_range)

model.plot_order_performance(order_performance_df)
```

```python
# wheat_us_hrw
df1 = dl.get_commodity_historical_monthly_data('wheat_us_hrw').dropna().reset_index(drop=True)
df1.columns = ['date', 'wheat_us_hrw']

model1 = ARIMAForecaster(df=df1.copy(), column='wheat_us_hrw', date_column='date', 
                         order=(1,0,0))
order_performance_df = model1.evaluate_ar_values(start_date='2018-01-01', 
                                                end_date='2019-12-01',
                                                order_range=order_range)

model1.plot_order_performance(order_performance_df)
```

```python
# wheat_canadi
df2 = dl.get_commodity_historical_monthly_data('canadi').dropna().reset_index(drop=True)
df2.columns = ['date', 'wheat_canadi']

model2 = ARIMAForecaster(df=df2.copy(), column='wheat_canadi', date_column='date', 
                        order=(1,0,0))

# for canadi, there is no data available after 2012-02-01
order_performance_df = model2.evaluate_ar_values(start_date='2011-01-01', 
                                                end_date='2011-12-01',
                                                order_range=order_range)

model2.plot_order_performance(order_performance_df)
```
